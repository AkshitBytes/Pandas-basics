#### PivotTables

```python
import numpy as np 
import pandas as pd 
```

```python
data = {
    'Date': pd.date_range('2023-01-01', periods=20),
    'Product': ['A', 'B', 'C', 'D'] * 5,
    'Region': [
        'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West',
        'North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'
    ],
    'Sales': np.random.randint(100, 1000, 20),
    'Units': np.random.randint(10, 100, 20),#start stop size
    'Rep': [
        'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary',
        'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice', 'John', 'Mary', 'Bob', 'Alice'
    ]
}

df = pd.DataFrame(data)
df['Month'] = df['Date'].dt.month_name()
df['Quarter'] = 'Q' + df['Date'].dt.quarter.astype(str)
```

```python
df # date is in year-month-date format
```

```python
#grouping data first
#there is a method set_index
#default aggregation is mean()
pd.pivot_table(df,values='Sales',index='Region',columns='Product',aggfunc='sum')
```

```python
# Pivot table with multiple values
pd.pivot_table(df,values=['Sales','Units'],index='Region',columns='Product')
```

#### Cross tabs

```python
#difference between pivot table and cross tabs is that pivot table uses aggregation functions and cross tabs use counting functions
pd.crosstab(df['Region'],df['Product'])
#general syntax : pd.crosstab(index, columns)
```
