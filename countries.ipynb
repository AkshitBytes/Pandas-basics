````python
if 'population' in df.columns:
    print('column exist')
else:
    print("Column does not exisst")

column_index = df.columns.get_loc('population')
print(column_index) 
#here i checked if a column names population exist or not if yes then at what index
````

````python
#1. which country has the least population
df.sort_values(by='population',ascending=True).head(1)[['country','population']]
````

````python
#2. what is the capital of the country with the least population
df.sort_values(by='population',ascending=True).head(1)[['country','capital_city','population']]
````

````python
#3.give me top 5 countries with highest democratic score
df.sort_values(by='democracy_score',ascending=False).head()[['country','democracy_score']]
````

````python
#4.how many total regions are there
# df['region'].nunique() #this is to find unique values in a data frame
#if you want to see the regions also you can
df['region'].unique()
````

````python
#5. how many countries lie in eastern europe region
mask = df['region'] == 'Eastern Europe'
df.loc[mask,['country']] #loc yses square brackets
````

````python
#6. who is the political leader of second highest populated country
df.sort_values(by='population',ascending=False).iloc[1][['country','political_leader']]
````

````python
#7.how many countries are there who's political leaders are unknown
mask = df['political_leader'].isnull() # isnull check for null values 
df.loc[mask,['country']]
````

````python
#8.how many countries have republic in there name
# mask = df['country'].str.contains('Republic')
# better way to do it is see for edge cases also
mask = df['country'].str.contains('Republic',case=False,na=False) #checks for republic also and handles countries with nan also
df.loc[mask,['country']]
````

````python
#9.which country in the africa region have the highest population
mask = df['region'].str.contains('Africa')
df_africa = df[mask]
df_africa.sort_values(by='population',ascending=False).head(1)[['country','population','region']]
````
